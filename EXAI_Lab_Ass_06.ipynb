```python
import os
import io
import math
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder, OneHotEncoder
from sklearn.impute import SimpleImputer, KNNImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.metrics import (accuracy_score, precision_score, recall_score, f1_score,
                             roc_auc_score, confusion_matrix, classification_report,
                             roc_curve, auc)

from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as ImbPipeline

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
import xgboost as xgb

import joblib

# Keras / TensorFlow
import tensorflow as tf
from tensorflow.keras import layers, models, Input, Model

# XAI
import shap
from lime import lime_tabular
from pdpbox import pdp, info_plots

GITHUB_RAW = "https://raw.githubusercontent.com/ArjunAnilPillai/Chronic-Kidney-Disease-dataset/main/final.csv"
LOCAL_FNAME = "final.csv"

def load_ckd():
    if os.path.exists(LOCAL_FNAME):
        print(f"Loading local {LOCAL_FNAME}")
        df = pd.read_csv(LOCAL_FNAME)
        return df
    else:
        try:
            print("Attempting download from GitHub mirror...")
            df = pd.read_csv(GITHUB_RAW)
            df.to_csv(LOCAL_FNAME, index=False)
            print("Downloaded and saved as", LOCAL_FNAME)
            return df
        except Exception as e:
            print("Download failed:", e)
            raise FileNotFoundError("Could not download dataset. Please download 'final.csv' from the GitHub repo and save in working directory.")

df = load_ckd()
print("Dataset shape:", df.shape)
display(df.head())

# -------------------------
# Quick EDA
# -------------------------
print("\n--- Dataframe info ---")
display(df.info())

print("\n--- Missing value counts ---")
missing = df.isnull().sum().sort_values(ascending=False)
display(missing[missing>0])

print("\n--- Summary statistics ---")
display(df.describe(include='all').T)

# Visualize class balance
if 'Class' in df.columns:
    target_col = 'Class'
elif 'class' in df.columns:
    target_col = 'class'
else:
    # Try common names
    possible = [c for c in df.columns if c.lower() in ['class', 'ckd', 'ckdstatus', 'target']]
    target_col = possible[0] if possible else df.columns[-1]
print("Target column used:", target_col)
print(df[target_col].value_counts())
sns.countplot(x=target_col, data=df)
plt.title("Target class distribution")
plt.show()

# Numeric/discrete variable distributions (select numerics)
numerics = df.select_dtypes(include=[np.number]).columns.tolist()
print("Numeric columns:", numerics)
for c in numerics:
    plt.figure(figsize=(5,2.5))
    sns.histplot(df[c].dropna(), kde=False, bins=30)
    plt.title(c)
    plt.tight_layout()
    plt.show()

# Correlation heatmap for numeric features
plt.figure(figsize=(12,10))
sns.heatmap(df[numerics].corr(), annot=True, fmt=".2f", cmap='coolwarm', center=0)
plt.title("Correlation heatmap (numeric)")
plt.show()

# -------------------------
# Preprocessing
# -------------------------
cat_cols = df.select_dtypes(exclude=[np.number]).columns.tolist()
cat_cols = [c for c in cat_cols if c != target_col]
num_cols = [c for c in numerics if c != target_col]

print("Categorical cols:", cat_cols)
print("Numeric cols:", num_cols)

def normalize_target(s):
    if s.dtype == object:
        return s.replace({'ckd':1, 'notckd':0, 'yes':1, 'no':0, 'present':1, 'absent':0, '1':1, '0':0})
    return s

df[target_col] = normalize_target(df[target_col]).astype(int)

num_imputer = KNNImputer(n_neighbors=5)
cat_imputer = SimpleImputer(strategy='most_frequent')

df_num = pd.DataFrame(num_imputer.fit_transform(df[num_cols]), columns=num_cols)
df_cat = pd.DataFrame(cat_imputer.fit_transform(df[cat_cols]), columns=cat_cols)

df_clean = pd.concat([df_num, df_cat, df[target_col].reset_index(drop=True)], axis=1)
print("After imputation shape:", df_clean.shape)
display(df_clean.head())

df_encoded = pd.get_dummies(df_clean.drop(columns=[target_col]), drop_first=True)
X = df_encoded
y = df_clean[target_col].astype(int)

print("X shape after encoding:", X.shape)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, stratify=y, random_state=42)
print("Train/test sizes:", X_train.shape, X_test.shape)

scaler = StandardScaler()
X_train_scaled = pd.DataFrame(scaler.fit_transform(X_train), columns=X_train.columns, index=X_train.index)
X_test_scaled = pd.DataFrame(scaler.transform(X_test), columns=X_test.columns, index=X_test.index)

print("Original class counts (train):\n", y_train.value_counts())
sm = SMOTE(random_state=42)
X_res, y_res = sm.fit_resample(X_train_scaled, y_train)
print("Resampled class counts:", np.bincount(y_res))

# -------------------------
# Utility: evaluation function
# -------------------------
def evaluate_model(name, model, X_test, y_test, show_report=True):
    y_pred = model.predict(X_test)
    if hasattr(model, "predict_proba"):
        y_proba = model.predict_proba(X_test)[:,1]
    elif hasattr(model, "decision_function"):
        try:
            scores = model.decision_function(X_test)
            y_proba = 1/(1+np.exp(-scores))
        except Exception:
            y_proba = np.zeros(len(y_pred))
    else:
        y_proba = np.zeros(len(y_pred))
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, zero_division=0)
    rec = recall_score(y_test, y_pred, zero_division=0)
    f1 = f1_score(y_test, y_pred, zero_division=0)
    try:
        roc = roc_auc_score(y_test, y_proba)
    except Exception:
        roc = float('nan')
    print(f"{name} -> Acc: {acc:.4f}, Prec: {prec:.4f}, Rec: {rec:.4f}, F1: {f1:.4f}, ROC-AUC: {roc:.4f}")
    if show_report:
        print(classification_report(y_test, y_pred, zero_division=0))
        cm = confusion_matrix(y_test, y_pred)
        sns.heatmap(cm, annot=True, fmt='d')
        plt.title(f"Confusion matrix: {name}")
        plt.xlabel("Predicted")
        plt.ylabel("Actual")
        plt.show()
    return dict(name=name, accuracy=acc, precision=prec, recall=rec, f1=f1, roc_auc=roc)

# -------------------------
# Part 3: Machine Learning Models
# -------------------------
results = []

lr = LogisticRegression(max_iter=1000)
lr.fit(X_res, y_res)
results.append(evaluate_model("Logistic Regression", lr, X_test_scaled, y_test))

dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_res, y_res)
results.append(evaluate_model("Decision Tree", dt, X_test_scaled, y_test))

rf = RandomForestClassifier(n_estimators=200, random_state=42, n_jobs=-1)
rf.fit(X_res, y_res)
results.append(evaluate_model("Random Forest", rf, X_test_scaled, y_test))

svm = SVC(kernel='rbf', probability=True, random_state=42)
svm.fit(X_res, y_res)
results.append(evaluate_model("SVM (RBF)", svm, X_test_scaled, y_test))

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_res, y_res)
results.append(evaluate_model("KNN", knn, X_test_scaled, y_test))

xgb_clf = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
xgb_clf.fit(X_res, y_res)
results.append(evaluate_model("XGBoost", xgb_clf, X_test_scaled, y_test))

res_df = pd.DataFrame(results).set_index('name')
display(res_df.sort_values(by='f1', ascending=False))

# -------------------------
# Part 4: Deep Learning Models
# -------------------------
X_train_dl = X_res.values
y_train_dl = y_res
X_test_dl = X_test_scaled.values
y_test_dl = y_test.values

def compile_and_train(model, X_train, y_train, X_val, y_val, epochs=30, batch_size=32, verbose=0):
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    hist = model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=epochs, batch_size=batch_size, verbose=verbose)
    return hist

def build_mlp(input_dim):
    inp = Input(shape=(input_dim,))
    x = layers.Dense(64, activation='relu')(inp)
    x = layers.Dense(32, activation='relu')(x)
    out = layers.Dense(1, activation='sigmoid')(x)
    return Model(inp, out, name="MLP")

mlp = build_mlp(X_train_dl.shape[1])
compile_and_train(mlp, X_train_dl, y_train_dl, X_test_dl, y_test_dl, epochs=50, batch_size=32, verbose=0)
y_proba = mlp.predict(X_test_dl).ravel()
y_pred = (y_proba >= 0.5).astype(int)
results.append(dict(name="MLP", accuracy=accuracy_score(y_test_dl, y_pred),
                    precision=precision_score(y_test_dl, y_pred, zero_division=0),
                    recall=recall_score(y_test_dl, y_pred, zero_division=0),
                    f1=f1_score(y_test_dl, y_pred, zero_division=0),
                    roc_auc=roc_auc_score(y_test_dl, y_proba)))
print("MLP done")

# CNN, LSTM, CNN+LSTM, Autoencoder blocks unchanged (keep as in your code)...

# -------------------------
# Part 5: Explainable AI (XAI)
# -------------------------
# 1) Feature importance RF
importances = rf.feature_importances_
feat_imp = pd.Series(importances, index=X_train_scaled.columns).sort_values(ascending=False)[:30]
plt.figure(figsize=(8,6))
sns.barplot(x=feat_imp.values[:20], y=feat_imp.index[:20])
plt.title("Top 20 feature importances (RandomForest)")
plt.show()

# 2) SHAP for RandomForest
print("Computing SHAP for RandomForest (TreeExplainer)...")
explainer_rf = shap.TreeExplainer(rf)
sample_idx = np.random.choice(X_test_scaled.shape[0], min(100, X_test_scaled.shape[0]), replace=False)
shap_values_rf = explainer_rf.shap_values(X_test_scaled.iloc[sample_idx])

if isinstance(shap_values_rf, list):
    shap_values_to_plot = shap_values_rf[1]
else:
    shap_values_to_plot = shap_values_rf

shap.summary_plot(shap_values_to_plot, X_test_scaled.iloc[sample_idx],
                  show=True, feature_names=X_test_scaled.columns.tolist())

# 3) SHAP for Logistic Regression
print("Computing SHAP for Logistic Regression (LinearExplainer)...")
explainer_lr = shap.LinearExplainer(lr, X_train_scaled)
shap_values_lr = explainer_lr.shap_values(X_test_scaled.iloc[sample_idx])
if isinstance(shap_values_lr, list):
    shap_values_lr = shap_values_lr[0]
shap.summary_plot(shap_values_lr, X_test_scaled.iloc[sample_idx],
                  show=True, feature_names=X_test_scaled.columns.tolist())

# 4) SHAP for MLP (DeepExplainer or Kernel fallback)
print("Computing SHAP for MLP (DeepExplainer)...")
try:
    background_data = X_train_dl[np.random.choice(X_train_dl.shape[0], min(100, X_train_dl.shape[0]), replace=False)]
    explainer_mlp_deep = shap.DeepExplainer(mlp, background_data)
    sample_idx_deep = np.random.choice(X_test_dl.shape[0], min(100, X_test_dl.shape[0]), replace=False)
    shap_values_mlp_deep = explainer_mlp_deep.shap_values(X_test_dl[sample_idx_deep])
    if isinstance(shap_values_mlp_deep, list):
        shap_values_to_plot = shap_values_mlp_deep[0]
    else:
        shap_values_to_plot = shap_values_mlp_deep
    shap.summary_plot(shap_values_to_plot, X_test_scaled.iloc[sample_idx_deep],
                      show=True, feature_names=X_test_scaled.columns.tolist())
except Exception as e:
    print("DeepExplainer failed, falling back to KernelExplainer:", e)
    mlp_predict = lambda x: mlp.predict(x).ravel()
    bg = X_train_scaled.sample(n=min(50, X_train_scaled.shape[0]), random_state=42)
    ke = shap.KernelExplainer(mlp_predict, bg.values)
    sv = ke.shap_values(X_test_scaled.iloc[sample_idx].values, nsamples=100)
    if isinstance(sv, list):
        sv_to_plot = sv[0]
    else:
        sv_to_plot = sv
    shap.summary_plot(np.array(sv_to_plot), X_test_scaled.iloc[sample_idx],
                      show=True, feature_names=X_test_scaled.columns.tolist())

# LIME + PDP blocks unchanged...
```

```python
# 2) SHAP (continued) - Trying DeepExplainer for MLP

# For MLP, DeepExplainer is often preferred over KernelExplainer
# Requires a background dataset (e.g., a sample of the training data)
print("Computing SHAP for MLP (DeepExplainer)...")

try:
    # DeepExplainer requires a background dataset and the model
    # Use a small sample of the training data as background
    background_data = X_train_dl[np.random.choice(X_train_dl.shape[0], min(100, X_train_dl.shape[0]), replace=False)]

    # Ensure background_data is a tensor if needed by DeepExplainer
    # Depending on shap version and tensorflow version, this might be necessary
    # background_data = tf.constant(background_data, dtype=tf.float32)


    explainer_mlp_deep = shap.DeepExplainer(mlp, background_data)
    # Sample a smaller subset of test data for speed if needed
    sample_idx_deep = np.random.choice(X_test_dl.shape[0], min(100, X_test_dl.shape[0]), replace=False)
    shap_values_mlp_deep = explainer_mlp_deep.shap_values(X_test_dl[sample_idx_deep])

    # shap_values for binary classification with DeepExplainer is a list of two arrays
    # We need the values for the positive class (index 1)
    # Ensure we are selecting the correct part for binary classification
    if isinstance(shap_values_mlp_deep, list):
        shap_values_to_plot = shap_values_mlp_deep[1] # Assuming index 1 is the positive class
    else:
        shap_values_to_plot = shap_values_mlp_deep # For single output models

    shap.summary_plot(shap_values_to_plot, X_test_scaled.iloc[sample_idx_deep], show=True, feature_names=X_test_scaled.columns.tolist())
    print("SHAP plot for MLP (DeepExplainer) generated.")

except Exception as e:
    print(f"Error computing or plotting SHAP values for MLP with DeepExplainer: {e}")
    print("Falling back to previous KernelExplainer attempt (which may still have issues).")
    # Keep the previous KernelExplainer code for MLP with the try-except block
    try:
        print("Computing SHAP for MLP (KernelExplainer - small sample)...")
        mlp_predict = lambda x: mlp.predict(x).ravel()
        bg = X_train_scaled.sample(n=min(50, X_train_scaled.shape[0]), random_state=42)
        ke = shap.KernelExplainer(mlp_predict, bg.values)
        sv = ke.shap_values(X_test_scaled.iloc[sample_idx].values, nsamples=100)
        # shap.summary_plot requires array shaped appropriately
        # Add an explicit check and selection for binary classification if sv is a list
        if isinstance(sv, list):
             sv_to_plot = sv[0] # For binary classification, KernelExplainer often returns a list with one array
        else:
             sv_to_plot = sv
        shap.summary_plot(np.array(sv_to_plot), X_test_scaled.iloc[sample_idx], show=True)
        print("SHAP plot for MLP (KernelExplainer) generated.")
    except Exception as e_kernel:
        print(f"Error plotting SHAP values for MLP with KernelExplainer: {e_kernel}")
```
